```python
# imports
import pandas as pd
```

##### importing data, cleaning, merging, etc. before creating visualization

```python
freedom_df = pd.read_csv('heritage-index-of-economic-freedom-2026-04-19_1352.csv', skiprows=4)
freedom_df = freedom_df.dropna(subset='Overall Score')
freedom_df
```

```python
opinion_df = pd.read_csv('AIopinions.csv')
opinion_df = opinion_df.dropna(how='all')

opinion_df
```

```python
x = opinion_df['Category'] == 'My job' # removes duplicates from the list

country_list = opinion_df.Country[x].tolist()

# after doing this once, the UK and Turkey were removed because they had different names in the different sets of data
    # im going to manually add the two names so they're not ommitted

country_list.append('United Kingdom')
country_list.append('Türkiye')
```

```python
# filtering freedom score data for only countries I want

freedom_df = freedom_df[freedom_df['Country'].isin(country_list)]
freedom_df.Country.size
```

```python
len(opinion_df[opinion_df['Category'] == 'My job'])
```

```python
freedom_df = freedom_df.reset_index()
freedom_df
```

```python
# now I just need to create a global average economic freedom score, then i'll have 33 rows to merge to the other 33 rows by their name

avg_freedom = freedom_df['Overall Score'].mean()
avg_freedom.round(1)
```

```python
freedom_df.loc[32] = {
    'Country': 'Global',
    'Index Year': 2024,
    'Overall Score': avg_freedom.round(1)
}
```

```python
freedom_df = freedom_df[['Country' , 'Overall Score']]
freedom_df
```

```python
freedom_df['Country'].replace('Türkiye' , 'Turkey' , inplace= True)
opinion_df['Country'].replace('Great Britain' , 'United Kingdom' , inplace= True) # I know they aren't the same, 
                                                                    # but the freedom score applies to the UK, as the UK is a subset of Great Britain

freedom_df
```

```python
# now I can merge

all_data = pd.merge(opinion_df , freedom_df , on = 'Country' , how='left')
all_data.rename(columns = {
    '% of respondents' : '% expecting improvement' , 
    'Overall Score' : 'Freedom Score', 
} , inplace = True)

all_data
```

```python
all_data["% expecting improvement"] = all_data["% expecting improvement"].str.replace("%","").astype(int)
all_data
```

```python
all_data['Freedom Score'].describe() # so I know the range to bin my values
```

```python
def binning(value):
    if pd.isna(value):
        raise ValueError("Freedom Score is missing, this should not happen")
    if value < 45:
        raise ValueError("Freedom Score is below 45, this should not happen (data min is 48.5)")
    elif value < 55:
        return "Low (45-55)"
    elif value < 65:
        return "Mid-low (55-65)"
    elif value < 75:
        return "Mid-high (65-75)"
    elif value < 85:
        return "High (75-85)"
    else: # >= 85
        raise ValueError("Freedom Score is 85 or higher, this should not happen (data max is 83.5)")


all_data['Freedom Score'] = all_data['Freedom Score'].apply(binning)
all_data
```

##### creating visualization

```python
# imports
import plotly.express as px
```

```python
# pivot table
df_plot = all_data.pivot(
    index=['Country', 'Freedom Score'],
    columns='Category',
    values='% expecting improvement'
).reset_index()

# rename to Economic Freedom Score everywhere
df_plot = df_plot.rename(columns={'Freedom Score': 'Economic Freedom Score'})
df_plot['OriginalBin'] = df_plot['Economic Freedom Score']

bin_list = ['All', 'Low (45-55)', 'Mid-low (55-65)', 'Mid-high (65-75)', 'High (75-85)']

# "All" frame
all_df = df_plot.copy()
all_df['Economic Freedom Score'] = 'All'

df_animated = pd.concat([all_df, df_plot], ignore_index=True)

# Duplicate global into every frame
global_base = df_plot[df_plot['Country'] == 'Global'].copy()
extra_globals = []
for b in bin_list:
    already_present = df_animated[
        (df_animated['Country'] == 'Global') & (df_animated['Economic Freedom Score'] == b)
    ]
    if already_present.empty:
        g = global_base.copy()
        g['Economic Freedom Score'] = b
        extra_globals.append(g)

if extra_globals:
    df_animated = pd.concat([df_animated] + extra_globals, ignore_index=True)

# color by original bin
df_animated['LegendGroup'] = df_animated.apply(
    lambda row: 'Global' if row['Country'] == 'Global' else row['OriginalBin'],
    axis=1
)

# NaN placeholders so Plotly clears missing groups per frame
all_legend_groups = ['Low (45-55)', 'Mid-low (55-65)', 'Mid-high (65-75)', 'High (75-85)', 'Global']
placeholder_rows = []
for b in bin_list:
    for lg in all_legend_groups:
        has_data = df_animated[
            (df_animated['Economic Freedom Score'] == b) &
            (df_animated['LegendGroup'] == lg)
        ]
        if has_data.empty:
            placeholder_rows.append({
                'Country': '__placeholder__',
                'Economic Freedom Score': b,
                'OriginalBin': lg,
                'LegendGroup': lg,
                'The job market': float('nan'),
                'My job': float('nan')
            })

if placeholder_rows:
    df_animated = pd.concat([df_animated, pd.DataFrame(placeholder_rows)], ignore_index=True)

# Custom group colors
color_discrete_map = {
    'Low (45-55)':      "#E07A5F",
    'Mid-low (55-65)':  "#F2CC8F",
    'Mid-high (65-75)': '#8DA9C4',
    'High (75-85)':     "#3D5A80",
    'Global':           'black'
}

# create figure
fig = px.scatter(
    df_animated,
    x='The job market',
    y='My job',
    color='LegendGroup',
    color_discrete_map=color_discrete_map,
    animation_frame='Economic Freedom Score',
    hover_name='Country',
    # Hide certain animation + legendgroup from tooltop
    hover_data={
        'Economic Freedom Score': False,
        'LegendGroup': False,
        'The job market': ':.1f',
        'My job': ':.1f'
    },
    category_orders={
        'LegendGroup': ['Low (45-55)', 'Mid-low (55-65)', 'Mid-high (65-75)', 'High (75-85)', 'Global'],
        'Economic Freedom Score': bin_list
    },
    labels={
        'The job market': '% Expecting AI Improvement - The Job Market',
        'My job': "% Expecting AI Improvement - Respondants' Specific Job",
        'LegendGroup': 'Group:'
    }
)

# formatting
fig.update_xaxes(range=[0, 70])
fig.update_yaxes(range=[0, 70])
fig.update_traces(marker=dict(size=14, line=dict(width=1, color='DarkSlateGrey')))

# Global = black diamond in base traces
fig.update_traces(
    marker=dict(size=18, color='black', symbol='diamond'),
    selector=dict(legendgroup='Global')
)

# apply diamond to every animation frame
for frame in fig.frames:
    for trace in frame.data:
        if trace.legendgroup == 'Global':
            trace.update(marker=dict(size=18, color='black', symbol='diamond'))

# informative textbox bottom right
fig.add_annotation(
    text=(
        "<b>Economic Freedom Score</b> measures how market-oriented<br>"
        "a country's economy is using factors such as property rights,<br>"
        "business regulation, trade openness, and government<br>"
        "involvement in the economy.<br><br>"
        "<b>Higher scores</b> generally indicate economies with stronger<br>"
        "private enterprise and more open markets.<br>"
        "<b>Lower scores</b> generally indicate greater government control<br>"
        "or state direction of economic activity."
    ),
    xref='paper', yref='paper',
    x=1.15, y=0.15,
    xanchor='right', yanchor='bottom',
    showarrow=False,
    font=dict(size=10, color='#333333'),
    bgcolor='rgba(255, 255, 255, 0.85)',
    bordercolor='#AAAAAA',
    borderwidth=1,
    borderpad=8,
    align='left'
)

# ensure all title colors are black
fig.update_layout(
    title=dict(
        text="AI Optimism: Respondant's Job vs. The Job Market<br>Filtered by Economic Freedom Score",
        x=0.5,
        xanchor='center',
        font=dict(color='black')     
    ),
    xaxis=dict(
        title=dict(
            text='% Expecting AI Improvement - The Job Market',
            font=dict(color='black') 
        )
    ),
    yaxis=dict(
        title=dict(
            text="% Expecting AI Improvement - Respondants' Specific Job",
            font=dict(color='black')  
        )
    ),
    legend_title='Group:',
    height=680,
    width=950,
    template='plotly_white',
    legend=dict(
        bordercolor='#AAAAAA',
        borderwidth=1,
        bgcolor='rgba(255, 255, 255, 0.85)',
        font=dict(color='black'),
        title=dict(
            text='Group:',
            font=dict(color='black')
        )
))

fig.show()

fig.write_html('ai_optimism_scatterplot.html')
fig.write_image('ai_optimism_scatterplot.png')
```
